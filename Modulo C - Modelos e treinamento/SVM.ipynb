```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

spotify = pd.read_csv('/content/drive/MyDrive/TCC Dataset/spotify_songs_2010_2019_normalized.csv')

df_pop = spotify[spotify['genre'].str.contains('^pop,', regex=True)]
df_hip_hop = spotify[spotify['genre'].str.contains('^hip hop,', regex=True)]

df_pop['genre'] = 'pop'
df_hip_hop['genre'] = 'hip hop'

df = pd.concat([df_pop, df_hip_hop])
```

```python
del df['artist']
del df['song']
# del df['duration_ms']
# del df['explicit']
del df['year']
# del df['popularity']
# del df['danceability']
# del df['energy']
# del df['key']
# del df['loudness']
# del df['mode']
# del df['speechiness']
# del df['acousticness']
# del df['instrumentalness']
# del df['liveness']
# del df['valence']
# del df['tempo']
# del df['genre']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

label_enc = LabelEncoder()

X = df.iloc[:,0:len(df.columns) - 1]
y = label_enc.fit_transform(df.iloc[:,-1])

X['explicit'] = LabelEncoder().fit_transform(df.iloc[:,1:2])

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, scale
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline

clf = Pipeline(steps=[('standardscaler', StandardScaler()), ('svc', SVC(random_state=42))])
model = clf.fit(X_train, y_train)
```

```python
y = model.predict(X_test)
y2 = model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
#accuracy training
accuracy_score(y, y_test)
```

```python
#accuracy test
accuracy_score(y_train, y2)
```

```python
from sklearn.metrics._plot.confusion_matrix import confusion_matrix
confusion = confusion_matrix(y_test, y)

print(confusion)
```

###GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.5, 1, 100],
    'gamma': ['scale', 1, 0.001],
    'kernel': ['poly', 'rbf']
}

opt_params = GridSearchCV(
    SVC(),
    param_grid,
    cv=5,
    scoring='accuracy',
    verbose=0
)

opt_params.fit(X_train, y_train)
print(opt_params.best_params_)
```
